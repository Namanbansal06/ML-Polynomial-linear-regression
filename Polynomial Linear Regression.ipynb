```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

# Loading data

```python
data = pd.read_csv("Resources/Position_Salaries.csv")
data.head()
```

```python
data.isnull().sum()
```

# Splitting the data

```python
x = data.iloc[:, 1:-1].values
y = data.iloc[:, -1].values
```

# Linear Regression Model

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(x,y)
```

# Visuliase the LR model

```python
plt.scatter(x,y,color = 'red' )
plt.plot(x, lin_reg.predict(x), color = 'blue')
plt.xlabel("input")
plt.ylabel("prediction")
plt.show()
```

# Polynomial Linear Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree =5)
x_poly = poly_reg.fit_transform(x)
```

```python
x_poly
```

```python
poly_lr = LinearRegression()
```

```python
poly_lr.fit(x_poly, y)
```

# Visulising Poly Linear Regression

```python
# degree =1
plt.scatter(x,y ,color = 'red')
plt.plot(x,poly_lr.predict(poly_reg.fit_transform(x)), color = 'blue')
plt.xlabel("input")
plt.ylabel("predctions")
plt.show()
```
